## Decision Tree for Full Data Set

```python
from sklearn.model_selection import train_test_split
# split data into testing and training sets
X = temp
y = pleasant
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=5)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
from matplotlib.pyplot import figure
from sklearn import tree
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
```

```python
# Use grid search to find parameters for decision tree
# Define parameter grid
param_grid = {
    'max_depth': [5, 10, 15, 20],
    'min_samples_split': [2, 5, 8, 11],
    'min_samples_leaf': [5, 10, 15, 20],
    'max_leaf_nodes': [None, 100, 600, 1000]
}

# Perform grid search with cross-validation
grid_search = GridSearchCV(DecisionTreeClassifier(criterion='gini'), param_grid, cv=3, scoring='accuracy', n_jobs=-2)
grid_search.fit(X_train, y_train)

# Best parameters from grid search
best_params = grid_search.best_params_
print("Best parameters for pruning:", best_params)

# Fit the pruned decision tree with the best parameters
climate_dt_best = grid_search.best_estimator_
climate_dt_best.fit(X_train, y_train)
```

```python
# Evaluate the best pruned model on the testing data
y_test_pred_best = climate_dt_best.predict(X_test)
test_accuracy_best = accuracy_score(y_test, y_test_pred_best)
print("Testing accuracy after best pruning:", test_accuracy_best)
```

```python
# Evaluate the best pruned model on the training data
y_train_pred_best = climate_dt_best.predict(X_train)
train_accuracy_best = accuracy_score(y_train, y_train_pred_best)
print("Training accuracy after best pruning:", train_accuracy_best)
```

```python
# plot the tree
figure(figsize=(15,15))
tree.plot_tree(climate_dt_best)
```

## Create Decision Tree for Heathrow Data Only

```python
# create table with Heathrow data only
Heathrow = temp[[col for col in temp.columns if 'HEATHROW' in col]]
Heathrow.head()
```

```python
# create Pleasantness table for Heathrow data only
Heathrow_pleasant = pleasant[['HEATHROW_pleasant_weather']]
Heathrow_pleasant.head()
```

```python
# split data into testing and training sets
X1 = Heathrow
y1 = Heathrow_pleasant
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.25, random_state=5)
print(X1_train.shape)
print(X1_test.shape)
print(y1_train.shape)
print(y1_test.shape)
```

```python
# Use grid search to find parameters for decision tree
param_grid = {
    'max_depth': [5, 10, 15, 20],
    'min_samples_split': [2, 5, 8, 11],
    'min_samples_leaf': [1, 2, 5, 10],
    'max_leaf_nodes': [None, 100, 300, 500]
}

# Perform grid search with cross-validation
grid_search = GridSearchCV(DecisionTreeClassifier(criterion='gini'), param_grid, cv=5, scoring='accuracy')
grid_search.fit(X1_train, y1_train)

# Best parameters from grid search
best_params = grid_search.best_params_
print("Best parameters for pruning:", best_params)

# Fit the pruned decision tree with the training data
Heathrow_dt_best = grid_search.best_estimator_
Heathrow_dt_best.fit(X1_train, y1_train)


# Evaluate the best pruned model on the testing data
y1_test_pred_best = Heathrow_dt_best.predict(X1_test)
test_accuracy_best = accuracy_score(y1_test, y1_test_pred_best)
print("Testing accuracy after best pruning:", test_accuracy_best)
```

```python
# Evaluate the best pruned model on the training data
y_train_pred_best = Heathrow_dt_best.predict(X_train)
train_accuracy_best = accuracy_score(y_train, y_train_pred_best)
print("Training accuracy after best pruning:", train_accuracy_best)
```

```python
# Plot tree
plt.figure(figsize=(15, 15))  
tree.plot_tree(y_train_pred_best, filled=True)
plt.show() 
```

```python
# determine the accuracy of the updated tree
y1_pred = Heathrow_dt_best.predict(X1_test)
y1_pred_train = Heathrow_dt_best.predict(X1_train)

accuracy_test = accuracy_score(y1_test, y1_pred)
accuracy_train = accuracy_score(y1_train, y1_pred_train)

print(f"Accuracy, Test: {accuracy_test:.2f}")
print(f"Accuracy, Train: {accuracy_train:.2f}")
```

```python
# create a confusion matrix for the Heathrow testing data
cm = confusion_matrix(y1_test, y1_pred)
plt.figure(figsize=(8, 6))
ax = sns.heatmap(cm, annot=True, fmt="d", linewidths=.5, cmap='Blues')
plt.title('Confusion Matrix for Heathrow')
plt.ylabel('Actual Pleasantness')
plt.xlabel('Predicted Pleasantness')
ax.set_xticklabels(['Unpleasant', 'Pleasant'])
ax.set_yticklabels(['Unpleasant', 'Pleasant'])
plt.show()
```
