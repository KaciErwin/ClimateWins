## Create Decision Tree for Full Data Set

```python
from sklearn.model_selection import train_test_split
# split data into testing and training sets
X = temp
y = pleasant
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=5)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# import libraries for Decision Tree classifier and run with climate data
from sklearn.tree import DecisionTreeClassifier
from matplotlib.pyplot import figure
from sklearn import tree

climate_dt = DecisionTreeClassifier(criterion='gini', min_samples_split=2)
climate_dt.fit(X_train, y_train)
figure(figsize=(15,15))
tree.plot_tree(climate_dt)
```

```python
# determine the accuracy of the tree
from sklearn.metrics import accuracy_score

y_pred = climate_dt.predict(X_test)
y_pred_train = climate_dt.predict(X_train)

accuracy_test = accuracy_score(y_test, y_pred)
accuracy_train = accuracy_score(y_train, y_pred_train)

print(f"Accuracy, Test: {accuracy_test:.2f}")
print(f"Accuracy, Train: {accuracy_train:.2f}")
```

```python
# crossvalidation of tree performance, 5-fold
from sklearn.model_selection import cross_val_score

accuracy_cv_train = cross_val_score(climate_dt, X_train, y_train, cv=5, scoring='accuracy')

print(f"Cross-Validated Accuracy, Train: {np.mean(accuracy_cv_train):.2f}")
```

```python
# Get the depth of the tree
tree_depth = climate_dt.get_depth()
print("Depth of the tree:", tree_depth)

# Get the number of nodes in the tree
tree_nodes = climate_dt.tree_.node_count
print("Number of nodes in the tree:", tree_nodes)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
# Create Confusion Matrix for testing data 
locations = list({x.split("_")[0] for x in pleasant.columns})
locations.sort()

key = "_pleasant_weather"
figure, ax = plt.subplots(3, 5, figsize=(25, 15))  
labels = ['unpleasant', 'pleasant']  
count = 0

for i in range(3):
    for j in range(5):
        if count < len(locations):  
            name = locations[count]
            cm = confusion_matrix(y_test.loc[:,name+key], y_pred[:,count])            
            sns.heatmap(cm, annot=True, fmt='g', ax=ax[i, j])
            ax[i, j].set_xlabel('Predicted labels')
            ax[i, j].set_ylabel('True labels')
            ax[i, j].set_title(name + ' Matrix')
            if cm.shape == (2, 2): 
                ax[i, j].xaxis.set_ticklabels(labels)
                ax[i, j].yaxis.set_ticklabels(labels)
            count += 1

plt.tight_layout()
plt.show()
```

```python
# Create Confusion Matrix for training data 
locations = list({x.split("_")[0] for x in pleasant.columns})
locations.sort()

y_pred_train = climate_dt.predict(X_train)

key = "_pleasant_weather"
figure, ax = plt.subplots(3, 5, figsize=(25, 15))  
labels = ['unpleasant', 'pleasant']  
count = 0

for i in range(3):
    for j in range(5):
        if count < len(locations):  
            name = locations[count]
            cm = confusion_matrix(y_train.loc[:,name+key], y_pred_train[:,count])            
            sns.heatmap(cm, annot=True, fmt='g', ax=ax[i, j])
            ax[i, j].set_xlabel('Predicted labels')
            ax[i, j].set_ylabel('True labels')
            ax[i, j].set_title(name + ' Matrix')
            if cm.shape == (2, 2): 
                ax[i, j].xaxis.set_ticklabels(labels)
                ax[i, j].yaxis.set_ticklabels(labels)
            count += 1

plt.tight_layout()
plt.show()
```

```python
# Use grid search to find parameters for decision tree
from sklearn.model_selection import GridSearchCV

# Define parameter grid for pruning
param_grid = {
    'max_depth': [15, 20, 25, 34],
    'min_samples_split': [2, 5, 8, 11],
    'min_samples_leaf': [1, 2, 5, 10],
    'max_leaf_nodes': [None, 100, 600, 1000]
}

# Perform grid search with cross-validation
grid_search = GridSearchCV(DecisionTreeClassifier(criterion='gini'), param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)

# Best parameters from grid search
best_params = grid_search.best_params_
print("Best parameters for pruning:", best_params)

# Fit the pruned decision tree with the best parameters
climate_dt_best = grid_search.best_estimator_
climate_dt_best.fit(X_train, y_train)

# Evaluate the best pruned model on the testing data
y_test_pred_best = climate_dt_best.predict(X_test)
test_accuracy_best = accuracy_score(y_test, y_test_pred_best)
print("Testing accuracy after best pruning:", test_accuracy_best)
```
